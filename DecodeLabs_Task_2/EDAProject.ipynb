<span style="color:#ff0000; font-weight:bold;">1- IMPORTS:</span>

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

<span style="color:#ff0000; font-weight:bold;">2- DATA OVERVIEW:</span>

```python
# Load & Read Dataset: 
data = pd.read_excel("cleaned_dataset.xlsx")
data.head(10)
```

```python
# Columns & data types:
print(data.shape)
print(data.columns)
data.info()
```

```python
# Descriptive Statistics:
data.describe()
```

```python
data.describe(include="object")
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The dataset contains 1,200 complete orders with no missing values, and each order has a unique OrderID and TrackingNumber. On average, customers purchase about 3 items per order, spending approximately 1,054 per order. Among the seven available products, **Printer** is the most frequently purchased. **Online** is the most commonly used payment method, **Cancelled** is the most common order status, **FREESHIP** is the most popular coupon code, and **Instagram** is the main source of customer referrals. The dataset covers the period from January 2023 to June 2025, with unit prices ranging from 11.39 to 699.93, reflecting a diverse range of products and order values.


<span style="color:#ff0000; font-weight:bold;">3- NUMERICAL DISTRIBUTIONS:</span>

```python
num_cols = data.select_dtypes(include=np.number).columns

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(data[col], kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The numerical features have different patterns. UnitPrice is spread quite evenly across the dataset, while TotalPrice is concentrated around lower values with a few much higher purchases. This suggests that most customers place relatively small orders.

<span style="color:#ff0000; font-weight:bold;">4- OUTLIERS VISUALISATION:</span>

```python
for col in num_cols:
    plt.figure(figsize=(4,2))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The boxplots show that most variables do not have noticeable outliers. However, TotalPrice contains a small number of unusually high values. These may represent large customer purchases rather than data errors.

<span style="color:#ff0000; font-weight:bold;">5- CORRELATION MATRIX:</span>

```python
plt.figure(figsize=(10,8))
sns.heatmap(data[num_cols].corr(),
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Matrix")
plt.show()
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
There is a clear positive relationship between UnitPrice and TotalPrice, which is expected because the total price increases as the unit price increases. Quantity and ItemsInCart are also positively related, showing that customers with more items in their carts often buy larger quantities.

<span style="color:#ff0000; font-weight:bold;">6-CATEGORICAL VARIABLES:</span>

```python
cat_cols = data.select_dtypes(include="object").columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=data, x=col)
    plt.xticks(rotation=45)
    plt.title(col)
    plt.show()
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The categorical variables are fairly balanced, with no category dominating the others. Products, payment methods, and order statuses are distributed across different categories, giving a good representation of customer behavior. Identifier columns such as OrderID, CustomerID, and TrackingNumber are unique and do not provide meaningful patterns for analysis.

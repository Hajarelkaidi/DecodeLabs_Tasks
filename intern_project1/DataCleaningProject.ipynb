<span style="color:#FF0000; font-weight:bold;">3- DATA QUALITY ASSESSEMENT :</span>

<p style="color:#2ca02c;"><b>a)- Check Missing Values :</b></p>

```python
        # Number of missing values: 
missing_values = data.isnull().sum()
print("** Missing values before cleaning are: **\n",missing_values)
        # Percentage of missing values:
per_missing_values = data.isnull().mean() * 100
print("\n** The percentages of missing values are: **\n",per_missing_values)
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The analysis of missing values showed that most of the dataset is complete, with no missing data in the majority of columns. However, the `CouponCode` column had 309 missing values, which is about 25.75% of the dataset. This suggests that a significant number of orders were made without using a coupon code, while all the other fields are fully filled.


<p style="color:#2ca02c;"><b>b)- Check Duplicates :</b></p>

```python
        # Search duplicate rows:
print("Duplicate rows:", data.duplicated().sum())
        # Search duplicate Order IDs:
print("Duplicate Order IDs:", data["OrderID"].duplicated().sum())
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The duplicate check showed that there are no repeated rows in the dataset, and no duplicate Order IDs were found. This means each order is unique, which helps ensure the reliability of the data for further analysis.


<p style="color:#2ca02c;"><b>c)- Check Unique Values :</b></p>

```python
categorical_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'CouponCode', 'ReferralSource']

for col in categorical_cols:
    print(f"\n{col}:")
    print(data[col].unique())
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The categorical columns were reviewed to check the unique values and make sure there were no inconsistencies. Overall, the categories were clean and consistent, with no spelling mistakes or formatting issues detected. The only missing values appeared in the `CouponCode` column, which simply indicates orders where no discount coupon was used.


<p style="color:#2ca02c;"><b>d)- Check Business Rules :</b></p>

```python
# Quantity:
invalid_quantity = data[data['Quantity'] <= 0]
print("Invalid Quantity:", invalid_quantity.shape[0])

# UnitPrice:
invalid_price = data[data['UnitPrice'] <= 0]
print("Invalid UnitPrice:", invalid_price.shape[0])

#ItemsInCart:
invalid_cart = data[data['ItemsInCart'] < 1]
print("Invalid ItemsInCart:", invalid_cart.shape[0])

#TotalPrice:
invalid_total = data[
    data['TotalPrice'].round(2) !=
    (data['Quantity'] * data['UnitPrice']).round(2)]
print("Invalid TotalPrice:", invalid_total.shape[0])
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
All the business rules were checked and everything was valid. The values in `Quantity`, `UnitPrice`, and `ItemsInCart` were all correct, and `TotalPrice` properly matched the calculation of Quantity × UnitPrice, with only minor differences due to decimal precision. No violations or inconsistencies were found in the data.


<p style="color:#2ca02c;"><b>e)- Outliers Detection :</b></p>

```python
numeric_cols = ["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]
for col in numeric_cols:
    # Calculate the first quartile (25%)
    Q1 = data[col].quantile(0.25)

    # Calculate the third quartile (75%)
    Q3 = data[col].quantile(0.75)

    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Filter rows where values are outside the normal range
    outliers = data[
        (data[col] < lower) |
        (data[col] > upper)
    ]
    # Print the number of outliers found in the column
    print(f"\n{col}: {outliers.shape[0]} outliers")
    # Display important details of the detected outliers
    print(outliers[["OrderID", "Product", "Quantity", "UnitPrice", "TotalPrice"]])
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The outlier analysis showed that `Quantity`, `UnitPrice`, and `ItemsInCart` do not contain any unusual values, which means the data in these columns is consistent. On the other hand, `TotalPrice` has 8 outliers, representing some unusually high transaction amounts. These cases may need further investigation to understand whether they are valid transactions or potential anomalies.


<span style="color:#FF0000; font-weight:bold;">4- DATA CLEANING & PREPARATION :</span>

```python
        # Fill missing CouponCode values:
data["CouponCode"] = data["CouponCode"].fillna("NOT USED")
        # Verify missing values:
print("Missing values after cleaning are:\n",data.isnull().sum())
```

<span style="color:#1f77b4;"><b><u>Comment:</u></b></span><br>
The missing values in the `CouponCode` column most likely mean that no coupon was applied during those purchases. To handle this, we will replace the missing values with "Not Used". This helps us keep the information clear and consistent, without losing any rows from the dataset.
